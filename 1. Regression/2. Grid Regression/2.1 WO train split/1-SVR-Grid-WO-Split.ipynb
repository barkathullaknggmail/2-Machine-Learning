```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid'],
              'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3,n_jobs=-1)
# fitting the model for grid search
grid.fit(independent, dependent)
```

```python
# print best parameter after tunng
# print(grid.best_params_)
re = grid.cv_results_
# print(re)
print("The r_value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age=float(input("AGE:"))
bmi=float(input("BMI:"))
children=float(input("Children:"))
gender=int(input("Gender Male 0 or 1:"))
smoker=int(input("Smoker Yes 0 or 1 :"))
```

```python
Prediction=grid.predict([[age,bmi,children,gender,smoker]])
print("Future Predocion{}".format(Prediction))
```
